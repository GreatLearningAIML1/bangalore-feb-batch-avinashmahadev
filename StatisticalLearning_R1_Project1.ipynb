```python
laliga = pd.read_csv("Laliga.csv",header=1)
type(laliga)
laliga.head()
laliga.describe()
laliga_names=pd.DataFrame(laliga.dtypes)
laliga_names.columns
laliga.shape
```

# Question 1

```python
print("**************Convert all dashes to 0 *******************")
laliga=laliga.replace("-", 0)

laliga_names=laliga_names.loc[laliga_names.index!='Team']
laliga_names=laliga_names.loc[laliga_names.index!='Debut']
laliga_names=laliga_names.loc[laliga_names.index!='Since/LastApp']
laliga_names= laliga_names.loc[laliga_names[0]=="object",:]
laliga_names
laliga[laliga_names.index] = laliga[laliga_names.index].apply(pd.to_numeric)
laliga.describe()
laliga.head(3)

```

# Quesion 2

```python
laliga["Debut"]=laliga["Debut"].str.slice(start=0,stop=4)
laliga["Debut"]=pd.to_numeric(laliga["Debut"])

laliga1980=laliga[(laliga["Debut"]>=1930) & (laliga["Debut"]<=1980)]
print("Teams That Debuted between 1930 and 1980")
laliga1980.groupby("Team")["Debut"].agg("min").sort_values()
```

# Question 3

```python
print("TOP 5 teams in terms of points")



laliga=laliga.sort_values(by="Points",ascending=False)

print("Top 5 teams with best winning percentage")
laliga[["Team","Points"]].head(5)
```

# Question 4

```python

def Goal_diff_count():
    Goal_diff_count =laliga["GoalsFor"] - laliga["GoalsAgainst"]
    Goal_diff_count.index=laliga["Team"]
    max_index=Goal_diff_count.idxmax()
    min_index=Goal_diff_count.idxmin()
    return Goal_diff_count,max_index,min_index
laliga4=Goal_diff_count()

print("Goal DIfference for all teams ", laliga4[0])
print("Team with Highest Goal Difference ", laliga4[1])
print("Team with Lowest Goal Difference ", laliga4[2])
```

# Question 5

```python
laliga["Winning_percent"]=(laliga["GamesWon"] / laliga["GamesPlayed"])*100
laliga=laliga.sort_values(by="Winning_percent",ascending=False)

print("Top 5 teams with best winning percentage")
laliga[["Team","Winning_percent"]].head(5)
```
